## 단어 맞히기 게임

````python
#@title Before we enter...
import random

word_set = ['word', 'card', 'msci', 'into', 'look', 'room', 'noon', 'take',
            'book', 'more', 'sure', 'nerd', 'cold', 'cool', 'bold', 'load',
            'mood', 'lane', 'line', 'core', 'wood', 'zoom', 'lazy', 'busy']
answer = random.choice(word_set)

hit = 0
trial = 1
guess = input('Input any character : ')
while hit < len(answer):
    hit_this = 0
    if trial > 1:
        guess_plus = input('Input another character : ')
        guess += guess_plus
    print('({})'.format(trial), end=' ')
    for i in range(len(answer)):
        if answer[i] in guess:
            print(answer[i], end=' ')
            hit_this += 1
        else:
            print('_', end=' ')
    print()
    hit = hit_this
    trial += 1
````

사용자에게 글자를 받아오는 방법은? 구글링!

````python
guess = input('Input any character :')
````

문자열 내에서 특정 문자의 위치 찾는 방법?

문자열.find() 를 써도 되지만, 답을 표시하기 편하게 노가다로 해보자.

````python
for i in range(len(answer)):    # 답의 길이 만큼 반복
    if answer[i] == guess:    # 받아온 글자가 해당 글자와 같다면
        print(answer[i], end=' ')    # 해당 위치에 글자 표시
    else:
        print('_', end=' ')    # 오답이면 밑줄만 표시
````

두번째 시도 부터는?

````python
guess_plus = input('Input another character :')
guess += guess_plus    # 받아온 단어 누적
print(guess)
````

````python
for i in range(len(answer)):
    if answer[i] in guess:    # 받아온 글자가 답에 포함되어 있다면
        print(answer[i], end=' ')
    else:
        print('_', end=' ')
````

다 맞힐 때 까지 반복

````python
hit = 0
trial = 1
guess = input('Input any character : ')    # 최초 글자 받아오기
while hit < len(answer):    # hit 수가 글자 수 보다 작을때만 실행
    hit_this = 0    # 정답 수 초기화
    if trial > 1:    # 두 번째 시기부터 실행
        guess_plus = input('Input another character : ')
        guess += guess_plus    # 받아온 단어 누적
    print('({})'.format(trial), end=' ')    # 몇 번째 시도인지 표시
    for i in range(len(answer)):    # 답의 길이 만큼 반복
        if answer[i] in guess:    # 받아온 글자가 답에 포함되어 있다면
            print(answer[i], end=' ')    # 해당 위치에 글자 표시
            hit_this += 1    # 정답 수 증가
        else:
            print('_', end=' ')    # 오답이면 밑줄만 표시
    print()    # 줄바꿈을 위해 공백 표시
    hit = hit_this    # hit 수 갱신
    trial += 1    # 시도 수 증가
````
