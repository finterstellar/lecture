```python
#@title Before we go...
import random
numbers = random.sample(range(10), 4)
answer = ''
for i in range(4):
    answer += str(numbers[i])

h,f,t = 0,0,1
while h<4:
    guess = input('Input any 4 digit number :')
    h,f = 0,0
    for i in range(len(answer)):
        if guess[i] in answer:
            if guess[i] == answer[i]:
                h += 1
            else:
                f += 1
    print('({})'.format(t), end=' ')
    if h == 4:
        print('Good job!')
    else:
        print(guess, ': {} hits, {} fouls'.format(h,f))
    t += 1
```
