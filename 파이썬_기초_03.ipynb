```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Create figure with secondary y-axis
fig = make_subplots(specs=[[{"secondary_y": True}]])

# Add traces
fig.add_trace(
    go.Scatter(x=[1, 2, 3], y=[40, 50, 60], name="yaxis data"),
    secondary_y=False,
)

fig.add_trace(
    go.Scatter(x=[2, 3, 4], y=[4, 5, 6], name="yaxis2 data"),
    secondary_y=True,
)

# Set y-axes titles
fig.update_yaxes(title_text="<b>primary</b> yaxis title", secondary_y=False)
fig.update_yaxes(title_text="<b>secondary</b> yaxis title", secondary_y=True)

fig.show()
```

```python
stocks = ['AAPL','MSFT','GOOGL','TSLA','META','F','KO']
df = pd.DataFrame()
for s in stocks:
    df_tmp = fdr.DataReader(s)
    df_tmp.rename(columns={'Close':s}, inplace=True)
    df = pd.concat([df,df_tmp[s]], axis=1)
df.fillna(method='ffill', inplace=True)
df.dropna(inplace=True)
dfr = df.iloc[-252:] / df.iloc[-252] * 100
dfr
```

```python
fig = go.Figure(data=[go.Surface(x=dfr.columns, y=dfr.index, z=dfr.values)])
fig.show()
```

[주말 감추기](https://plotly.com/python/time-series/#hiding-weekends-and-holidays)

```python
fig = px.area(df, x=df.index, y='AAPL')
fig.update_xaxes(
    rangebreaks=[
        dict(bounds=["sat", "mon"]), #hide weekends
        dict(values=["2022-12-25", "2021-01-01"])  # hide Christmas and New Year's
    ]
)
fig.show()
```
