```python
import finterstellar as fs
import pandas as pd
```

```python
fs.get_price_kr('삼성전자', start_date='2020-01-01')
```

```python
sec = fs.get_ohlc_kr('삼성전자',start_date='2022-01-01', end_date='2023-12-31')[['Close']]
sec.rename(columns={'Close':'SEC'}, inplace=True)
sec
```

```python
fs.get_ohlc('AAPL', start_date='2020-01-01')
```

```python
aapl = fs.get_price('AAPL', start_date='2022-01-01', end_date='2023-12-31')
aapl
```

```python
df = pd.concat([aapl, sec], axis=1)
df
```

```python
df.loc['2022-12-20':'2023-01-01']
```

```python
df.fillna(method='ffill', inplace=True)
df
```

```python
df.plot()
```

```python
df.iloc[0]
```

```python
# 지수화
df1 = df / df.iloc[0] * 100
df1.plot()
```

```python
# 일간 변동률
df2 = df.pct_change()
df2.plot()
```

```python
df2
```

```python
# 변동성
df2.std()
```

```python
import plotly.express as px
```

```python
fig = px.line(df1, x=df1.index, y=['AAPL','SEC'])
fig.show()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Create figure with secondary y-axis
fig = make_subplots(specs=[[{"secondary_y": True}]])

# Add traces
fig.add_trace(
    go.Scatter(x=df.index, y=df['AAPL'], name='Apple'),
    secondary_y=False,
)

fig.add_trace(
    go.Scatter(x=df.index, y=df['SEC'], name='삼성전자'),
    secondary_y=True,
)

# Set y-axes titles
fig.update_yaxes(title_text='Apple', secondary_y=False)
fig.update_yaxes(title_text='삼성전자', secondary_y=True)

fig.update_layout(
    template='plotly_dark',
    margin=dict(l=10, r=10, t=10, b=10),
    legend=dict(yanchor='top', y=0.99, xanchor='left', x=0.01)
)
fig.show()
```

```python
df['AAPL_MA5'] = df['AAPL'].rolling(5).mean()
df['SEC_MA5'] = df['SEC'].rolling(5).mean()
df.tail()
```

```python
# Create figure with secondary y-axis
fig = make_subplots(specs=[[{"secondary_y": True}]])

# Add traces
fig.add_trace(
    go.Scatter(x=df.index, y=df['AAPL_MA5'], name='Apple'),
    secondary_y=False,
)

fig.add_trace(
    go.Scatter(x=df.index, y=df['SEC_MA5'], name='삼성전자'),
    secondary_y=True,
)

# Set y-axes titles
fig.update_yaxes(title_text='Apple', secondary_y=False)
fig.update_yaxes(title_text='삼성전자', secondary_y=True)

fig.update_layout(
    template='plotly_dark',
    margin=dict(l=10, r=10, t=10, b=10),
    legend=dict(yanchor='top', y=0.99, xanchor='left', x=0.01)
)
fig.show()
```
