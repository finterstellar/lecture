## 파일 다운받아서 업로드 하기 귀찮다면

```python
import requests    # browser 와 server 간 통신
from io import StringIO    # 문자열을 파일처럼
```

```python
# url: 서버 주소
url = 'https://fred.stlouisfed.org/graph/fredgraph.csv'
```

```python
params = {
    'id': 'UNRATE',
}
```

```python
r = requests.get(url, params=params)
```

```python
r.text
```

```python
pd.read_csv(StringIO(r.text), index_col=0, parse_dates=True, header=0, na_values=".",)
```
