```python
#  
pd.options.display.float_format = '{:,.2f}'.format
pd.set_option('mode.chained_assignment', None)
```

```python
# url:  
url = 'https://opendart.fss.or.kr/api/corpCode.xml'
```

```python
payload = {
    'crtfc_key': '283aee3cde68b4f293f55ece4c9379c3eacd0e7f'
}
```

```python
zip = requests.get(url, params=payload).content
```

```python
from zipfile import ZipFile
from io import BytesIO
```

```python
z = ZipFile(BytesIO(zip))
```

```python
z.namelist()
```

```python
z.read('CORPCODE.xml').decode()
```

```python
cd = pd.read_xml(z.read('CORPCODE.xml').decode())
cd
```
