## 기업 고유번호

```python
import requests    # browser 와 server 간 통신
import pandas as pd    # 데이터 분석 도구
```

```python
# 데이터 포맷팅
pd.options.display.float_format = '{:,.2f}'.format
pd.set_option('mode.chained_assignment', None)
```

```python
# url: 서버 주소
url = 'https://opendart.fss.or.kr/api/corpCode.xml'
```

```python
payload = {
    'crtfc_key': '283aee3cde68b4f293f55ece4c9379c3eacd0e7f'
}
```

```python
zip = requests.get(url, params=payload).content
```

```python
from zipfile import ZipFile
from io import BytesIO
```

```python
z = ZipFile(BytesIO(zip))
```

```python
z.namelist()
```

```python
z.read('CORPCODE.xml').decode()
```

```python
cd = pd.read_xml(z.read('CORPCODE.xml').decode())
cd
```

```python
cd[cd['corp_name']=='삼성전자']
```

```python
cd[cd['corp_name']=='삼성전자']['corp_code'].iloc[0]
```
