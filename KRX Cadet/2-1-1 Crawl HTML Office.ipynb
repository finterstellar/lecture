```python
import requests
from bs4 import BeautifulSoup    # BeautifulSoup 모듈 호출
```

```python
url = 'http://book.finterstellar.com/sample.html'
```

```python
r = requests.get(url)
```

```python
r
```

```python
r.text
```

```python
soup = BeautifulSoup(r.text, 'lxml')
soup
```

```python
print(soup.prettify())
```

```python
soup.title
```

```python
soup.title.text
```

```python
soup.a
```

```python
soup.find_all('a')
```

```python
soup.find_all('a').text    # .text를 이용하려면 값을 지정해줘야함
```

```python
soup.find_all('a')[2].text
```

```python
member = soup.find_all('a')
for m in member:
    print(m.text)
```

```python
member
```

```python
members = []    # 데이터를 담을 리스트 선언
for m in member:
    members.append(m.text)    # 리스트에 데이터 추가
members
```
